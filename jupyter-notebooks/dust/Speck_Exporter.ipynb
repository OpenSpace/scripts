```python
from astropy.table import Table
import io
import sys
from astropy.io import ascii
import numpy

input_filename='datasets/als_ii_f_MW3D.fits'
t = Table.read(input_filename)

file_description = "als_ii_f_MW3D"
output_filename = "als_ii_f_MW3D"

#specify names of table columns that will be written out in the first three positions
xcol = 'X'
ycol = 'Y'
zcol = 'Z'
extraCol = 'RA'

#find all columns that are not xyz
#not_xyz_cols = [col for col in t.colnames if (col!=xcol and col!=ycol and col!=zcol)]

#resort column order so xyz in front
col_order = [xcol,ycol,zcol, extraCol]
t = t[col_order]

writeFile = open("output/" + output_filename + '.speck','w')
writeFile.write("# {}\n".format(file_description))

#for colnum,colname in enumerate(['RA']):
#    print(colname,colnum)
#    writeFile.write("datavar {} {}\n".format(colnum,colname))
writeFile.write("datavar {} {}\n".format(0,extraCol))
     
for row in t:
    if not numpy.isnan(row[xcol]) :
        writeFile.write(" {} {} {} {}\n".format(row[xcol],row[ycol],row[zcol],row[extraCol]))

#    stdout = io.StringIO()
#    sys.stdout = stdout
#    if not numpy.isnan(row[xcol]) :
        #grab row in string format. 
        #This doesn't actually write anything to disk (only to sys.stdout, which we grab from the variable stdout)
#        ascii.write(row, sys.stdout,format='no_header')

        #write row to file
#        writeFile.write(stdout.getvalue())

writeFile.close()
```

```python
#create asset lua
import random

colors = [random.uniform(0, 1),random.uniform(0, 1),random.uniform(0, 1)]


asset = ""
asset += "local " + output_filename + " = {\n"
asset += "\tIdentifier = '" + output_filename + "',\n"
asset += "\tParent = 'Root',\n"
asset += "\tRenderable = {\n"
asset += "\t\tType = 'RenderableBillboardsCloud',\n"
asset += "\t\tEnabled = true,\n"
asset += "\t\tColor = {"+str(colors[0])+","+str(colors[1])+","+str(colors[2])+"},\n"
asset += "\t\tOpacity = 1,\n"
asset += "\t\tFile = asset.localResource('" + output_filename + ".speck'),\n"
asset += "\t\tPolygonSides = 6,\n"
asset += "\t\tUnit = 'pc',\n"
asset += "\t\tScaleFactor = 385,\n"
asset += "\t\tBillboardMinMaxSize = { 0.0, 300.0 },\n"
asset += "\t\tEnablePixelSizeControl = false\n"
asset += "\t},\n"
asset += "\tGUI = {\n"
asset += "\t\t\tName = '" + output_filename + "',\n"
asset += "\t\t\tPath = '/MW3D'\n"
asset += "\t}\n"
asset += "}\n\n"


asset += "asset.onInitialize(function()\n"
asset += "\topenspace.addSceneGraphNode(" + output_filename + ")\n"
asset += "end)\n"

asset += "asset.onDeinitialize(function()\n"
asset += "\topenspace.removeSceneGraphNode(" + output_filename + ")\n"
asset += "end)\n\n"


asset += "asset.export(" + output_filename + ") "


asset_file = open("output/" + output_filename + ".asset", "w")
asset_file.write(asset)
asset_file.close()
```
