```python
import os
for root, dirs, files in os.walk("./data"):
    super_asset = ""
    for name in dirs:
        for _, _, files in os.walk(root+"/"+name):
            print(name)
            #create asset for this directory
            asset = "local assetHelper = asset.require('util/asset_helper')\n"
            asset += "local transforms = asset.require('scene/solarsystem/sun/transforms')\n"
            asset += "local nodes = {}\n"
            if (name == 'Grids'):
                continue
            for file in files:
                if (file.endswith('.speck')):
                    if ('ellips' in file):
                        print("Elipse file: " + file)
                        
                    else:
                        print("star speck: " + file)
                        asset += "local stars = {\n\
                          Identifier = '" + os.path.splitext(file)[0] + "',\n\
                          Renderable = {\n\
                            Type = 'RenderableStars',\n\
                            File = asset.localResource('data/" + name + "/" + file + "'),\n\
                            Texture = asset.localResource('halo.png'),\n\
                            ColorMap = asset.localResource('colorbv.cmap'),\n\
                            MagnitudeExponent = 6.3,\n\
                            SizeComposition = 'Distance Modulus',\n\
                            RenderMethod = 'Texture Based', -- or PSF\n\
                            DataMapping = {\n\
                              Bv = 'color',\n\
                              Luminance = 'lum',\n\
                              AbsoluteMagnitude = 'absmag',\n\
                              ApparentMagnitude = 'appmag',\n\
                              Vx = 'vx',\n\
                              Vy = 'vy',\n\
                              Vz = 'vz',\n\
                              Speed = 'speed'\n\
                            }\n\
                          },\n\
                          GUI = {\n\
                            Name = '" + os.path.splitext(file)[0] + "',\n\
                            Path = '/Other/BANYAN Sigma'\n\
                          }\n\
                        }\n"
                        asset += "table.insert(nodes, stars)\n"
            asset += "assetHelper.registerSceneGraphNodesAndExport(asset, nodes)\n"
            asset_file = open(name + ".asset", "w")
            asset_file.write(asset)
            asset_file.close()
            super_asset += "asset.require('./" + name + ".asset')\n"
    if (super_asset):
        super_asset_file = open("BANYANSigma.asset", 'w')
        super_asset_file.write(super_asset)
        super_asset_file.close()
```
