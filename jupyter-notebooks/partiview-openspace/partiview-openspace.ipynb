```python
import os, re
for root, dirs, files in os.walk("./data"):
    super_asset = ""
    for name in dirs:
        for _, _, files in os.walk(root+"/"+name):
            print(name)
            #create asset for this directory
            asset = "local assetHelper = asset.require('util/asset_helper')\n"
            asset += "local transforms = asset.require('scene/solarsystem/sun/transforms')\n"
            asset += "local nodes = {}\n"
            asset += "local PARSEC_CONSTANT = 3.0856776E16;\n"
            if (name == 'Grids'):
                continue
            for file in files:
                if (file.endswith('.speck')):
                    if ('ellips' in file):
                        #reading file
                        print("Elipse file: " + file)
                        path = os.path.realpath(file)
                        print(path)
                        specked = open("data/" + name + "/" + file,"r")
                        firstline = specked.readline()
                        
                        #reading ellipse first line
                        numbers = firstline.split()
                        snumbers = re.split('-r |-n ',firstline)
                        
                        #splitting rotation nums
                        rnumbers = re.split('-n 36,19 ',firstline)
                        rotation = re.split(" ",str(rnumbers[1]))
                        print(rotation[0:9])
                        
                        #splitting position numbers
                        position = [numbers[0],numbers[1],numbers[2]]
                        
                        radius = [numbers[9],numbers[10],numbers[11]]
                        
                        #print (snumbers)
                        #print (snumbers[1])
                        
                        scale = re.split(",",str(snumbers[1]))
                        #print(scale[0])
                        #print(scale[1])
                        #print(scale[2])
                        
                        asset += "local ellipseMesh  = {\n\
                          Identifier = '" + os.path.splitext(file)[0] + "',\n\
                          Renderable = {\n\
                              Type = 'RenderableSphericalGrid',\n\
                              Color = { 1.0, 0.5, 0.2 },\n\
                              LineWidth = 2.0,\n\
                              Segments = 40,\n\
                              Enabled = false\n\
                            },\n\
                          Transform = {\n\
                              Scale = { \n\
                                Type = 'NonUniformStaticScale', \n\
                                Scale = {" +  scale[0] + " * PARSEC_CONSTANT," +  scale[1] + " * PARSEC_CONSTANT," +  scale[2] + " * PARSEC_CONSTANT}\n\
                                },\n\
                              Rotation = {\n\
                                Type = 'StaticRotation',\n\
                                Rotation ={"+ rotation[0] +","+ rotation[1] +","+ rotation[2] +","+ rotation[3] +","+ rotation[4] +","+ rotation[5] +","+ rotation[6] +","+ rotation[7] +","+ rotation[8] +"}\n\
                                },\n\
                              Translation = {\n\
                                Type = 'StaticTranslation',\n\
                                Position = {"+ position[0] +" * PARSEC_CONSTANT,"+ position[1] +" * PARSEC_CONSTANT,"+ position[2] +" * PARSEC_CONSTANT}\n\
                                }\n\
                              },\n\
                              GUI = {\n\
                            Name = '" + os.path.splitext(file)[0] + "',\n\
                            Path = '/Other/Partiview Ellipses'\n\
                          }\n\
                        }\n" 
                        asset += "table.insert(nodes, ellipseMesh)\n"
                        
                        
                        
                    else:
                        print("star speck: " + file)
                        asset += "local stars = {\n\
                          Identifier = '" + os.path.splitext(file)[0] + "',\n\
                          Renderable = {\n\
                            Type = 'RenderableStars',\n\
                            File = asset.localResource('data/" + name + "/" + file + "'),\n\
                            Texture = asset.localResource('halo.png'),\n\
                            ColorMap = asset.localResource('colorbv.cmap'),\n\
                            MagnitudeExponent = 6.3,\n\
                            SizeComposition = 'Distance Modulus',\n\
                            RenderMethod = 'Texture Based', -- or PSF\n\
                            DataMapping = {\n\
                              Bv = 'color',\n\
                              Luminance = 'lum',\n\
                              AbsoluteMagnitude = 'absmag',\n\
                              ApparentMagnitude = 'appmag',\n\
                              Vx = 'vx',\n\
                              Vy = 'vy',\n\
                              Vz = 'vz',\n\
                              Speed = 'speed'\n\
                            }\n\
                          },\n\
                          GUI = {\n\
                            Name = '" + os.path.splitext(file)[0] + "',\n\
                            Path = '/Other/Partiview Stars'\n\
                          }\n\
                        }\n"
                        asset += "table.insert(nodes, stars)\n"
            asset += "assetHelper.registerSceneGraphNodesAndExport(asset, nodes)\n"
            asset_file = open(name + ".asset", "w")
            asset_file.write(asset)
            asset_file.close()
            super_asset += "asset.require('./" + name + ".asset')\n"
    if (super_asset):
        super_asset_file = open("BANYANSigma.asset", 'w')
        super_asset_file.write(super_asset)
        super_asset_file.close()
```
